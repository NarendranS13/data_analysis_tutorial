```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
## load the file
dirname = os.getcwd()
# print(dirname)
data_folder = "data"
filename = "sales_data.csv"
input_file = os.path.join(dirname,data_folder,filename)
# print(input_file)

sales = pd.read_csv(input_file)
```

Relationship between the columns?


```python
corr = sales.corr(numeric_only=True)
corr
```

```python
fig = plt.figure(figsize=(8,8))
plt.matshow(corr, cmap="RdBu", fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical');
plt.yticks(range(len(corr.columns)), corr.columns);
```

```python
sales["Revenue_per_Age"].plot(kind="density", figsize=(6,6))
```
