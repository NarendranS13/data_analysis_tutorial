Visualization Concepts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

Global API

```python
# Matplotlib default pyplot API has a global, MATLAB style interface, as we've already seen
## Global API -- MATAB Based graphing.
### We need to refer OOP API ()
x = np.arange(-10,11)

plt.figure(figsize=(12,6))
plt.title('My Nice Plot')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x **2))
```

```python
#### Build two visualization using Global API

plt.figure(figsize = (12,6))
plt.title('My Nice Plot')

## Create 1st subplot
plt.subplot(1,2,1)
plt.plot(x, x ** 2)
plt.plot([0,0,0],[-10,0,100])
plt.legend(['X^2','Vertical line'])
plt.xlabel('X')
plt.ylabel('X Squared')

## Create 2nd Subplot

plt.subplot(1,2,2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10,0,10],[-50,-50,-50])
plt.legend(['-X^2','Horizontal line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

OOP Interface

```python
### USING OOP (OBJEct oriented programmg) we can build an common fig, axes and reuse them
fig, axes = plt.subplots(figsize = (12,6))
```

```python
axes.plot(
    x, (x**2), color='red', linewidth=3,
    marker='o', markersize = 8, label='X^2'
)

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')
axes.set_ylabel('X Squared')
axes.set_title("My Nice plot")

axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize = (12,6))

axes.plot(x, x+0, linestyle = 'solid')
axes.plot(x, x+1, linestyle = 'dashed')
axes.plot(x, x+2, linestyle='dashdot')
axes.plot(x, x+3, linestyle='dotted')
```

Types of plots

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2 # 0 to 15 radii
```

```python
plt.figure(figsize=(12,6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

Bar Charts

```python
Y = np.random.rand(1,5)[0]
Y2 = np.random.rand(1,5)[0]
```

```python
plt.figure(figsize=(12,4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

plt.show()
```

```python
plt.figure(figsize=(12,4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', label='label Y2')

plt.legend()
plt.show()
```
