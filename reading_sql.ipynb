Reading SQL via Pandas

```python
import pandas as pd
import sqlite3
```

Accessing DB via traditional connection method

```python
conn = sqlite3.connect('data/sakila.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM actor LIMIT 5;")
```

```python
results = cur.fetchall()
```

```python
results
```

```python
df = pd.DataFrame(results)
```

```python
df
```

Using pandas read_sql method

```python
conn = sqlite3.connect('data/sakila.db')
```

```python
df = pd.read_sql('SELECT * from actor limit 10', conn)
```

```python
df
```

Using pandas read_sql_query method

```python
df = pd.read_sql_query('Select * from actor', conn, 
                       index_col = 'actor_id',
                       parse_dates = 'last_modified')
```

```python
# df
conn.close()
```
