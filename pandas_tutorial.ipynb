# Pandas Tutorial

Pandas Series


```python
### G7 Nations. Canada, France, Germany, Italy, Japan, UK and US.
import pandas as pd
import numpy as np

# In millions
g7_pop = pd. Series([35.467, 63.951, 80.940, 60.665, 127.06, 64.511, 318.523])
```

```python
g7_pop
```

```python
# This access the index of the series.
g7_pop[1]
g7_pop[7]
```

```python
g7_pop.index
```

But, in contrast to lists, we can change the index

```python
g7_pop.index = [
    "Canada",
    "France",
    "Germany",
    "Italy",
    "Japan",
    "United Kingdom",
    "United States",
]
```

```python
g7_pop
```

## Indexing

```python
g7_pop["Canada"]
```

```python
g7_pop['Japan']
```

Numeric position can be used with iloc attribute

```python
### Find the population of France with iloc
g7_pop.iloc[1]
```

```python
### Find the population of US with iloc
g7_pop.iloc[-1]
```

Selecting multiple index at once

```python
g7_pop[['United Kingdom','United States']]
```

Slicing also works, but important in Pandas the upper limit is also included

```python
g7_pop['Canada':'Italy']
```

Conditional Selection (Boolean Arrays)

```python
g7_pop > 70
```

```python
g7_pop[g7_pop > 70]
```

```python
g7_pop[g7_pop > g7_pop.mean()]
```

```python
g7_pop.mean()
```

Operations on Series

```python
g7_pop * 2
```

Logic OR & AND

```python
g7_pop[(g7_pop > 80) & (g7_pop < 200)]
```

Modifying Series

```python
g7_pop['Canada'] = 40.5
```

```python
g7_pop.iloc[-1] = 500
```

```python
g7_pop
```

```python
g7_pop[g7_pop < 70]
```

```python
g7_pop[g7_pop < 70] = 99.99
```

```python
g7_pop
```
