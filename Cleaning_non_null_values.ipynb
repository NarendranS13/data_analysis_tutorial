## Cleaning Non null values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Sex': ['M','F','F','D','?'],
    'Age':[29,30,24,290,25]
})
```

```python
df
```

Finding Unique Values

```python
df['Sex'].unique()
```

```python
df['Sex'].value_counts()
```

```python
### Simple replace

df.replace({
    'Sex': {
        '?':'F',
        'D':'M'
    },
    'Age':{
        290:29
    }
})
```

```python
### Pythonic way to handle age

df[df['Age'] > 100]
```

```python
df.loc[df['Age'] > 100, 'Age'] = df.loc[df['Age'] > 100, 'Age'] /10
```

```python
df
```

Duplicates

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany'
],  index = [

    'Gerard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Witing',
    'Peter Ammon',
    'Klaus Scharioth'
])
```

```python
ambassadors
```

```python
ambassadors.duplicated()

### When using duplicated, the first instance is not duplicate and subesquent record will be duplicate
```

```python
ambassadors.duplicated(keep='last')
```

```python
ambassadors.duplicated(keep=False)
```

```python
ambassadors.drop_duplicates()
```

```python
ambassadors.drop_duplicates(keep = 'last')
```

```python
ambassadors.drop_duplicates(keep=False)
```
